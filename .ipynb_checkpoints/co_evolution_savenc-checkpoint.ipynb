```python
figfolder = './'

#load data 

dataset = 'IMERG'
dri = '/lss/research/agon-lab/IMERG-daily/NASA_interpolated/equ_waves/'
fn = '1deg_IMERG_2000-2021_IG0.nc'
data_kw = xr.open_dataset(dri+fn)
data_kw
```

```python
#chosen base point NH latitude

lats = 5. 
latn = 5. 

# Chosen base point longitude
lonw = 250.
lone = 250. 

#chosen base point NH latitude
lat_SHs=-5
lat_SHn=-5

#chosen time

mn0 = 2
mnf = 4
yr0 = 2001
yrf = 2021
time0 = str(yr0)+'-01-01T00:00:00.000000000'
timef = str(yrf)+'-12-31T00:00:00.000000000'
```

```python
time_all1 = data_kw['time']
```

```python
# set up conditional statements 
cond1_fma = (time_all1['time.month'] >= mn0) & (time_all1['time.month'] <= mnf) # season of interest
cond2_fma = (time_all1['time.year'] <= yrf) & (time_all1['time.year'] >= yr0) # season of interest
# now find only the times when MJO amplitude is in season of interest AND exceeds some threshold
time_fma = time_all1.where((cond1_fma)& (cond2_fma), drop=True) 
```

```python
if (lats == latn) | (lonw == lone):
    pr1 = data_kw['precipitationCal'].sel(lat=latn,lon=lonw, time = time_fma)
    pr_FMA_std_lon = data_kw['precipitationCal'].sel(lat=latn,lon=slice(225,285),time = time_fma).mean('lon').std(dim='time')
    print('1 Base Point')
else:
    pr1 = data_kw['precipitationCal'].sel(lat=slice(lats,latn),lon=slice(lonw,lone), time = time_fma).mean(dim=('lat','lon'))
    pr_FMA_std_lon = data_kw['precipitationCal'].sel(lat=slice(lats,latn),lon=slice(225,285),time = time_fma).mean(dim=('lat','lon')).std(dim='time')
    print('Base Region')

std = pr1.std(dim=('time'))
pr1
```

```python
#set up condition to choose precip greater than 1 std

pr = pr1
cond = (pr >= std)
```

```python
#Subset of precip data when precip is greater than 1 std

time_strong = pr.where(cond, drop=True)
```

```python
time_NH = time_strong['time']
```

```python
pr_NH_lag0 = data_kw['precipitationCal'].sel(lat=slice(-20,20),lon=slice(225,285), time = time_NH).mean(dim=('time'))
```

```python
#Chosen latitude  & longitude window

latitude_s = 2.5
latitude_n = 7.5

longi_w= 100
longi_e= 359
```

```python
pr_avg = data_kw['precipitationCal'].sel(lat=slice(latitude_s,latitude_n),lon=slice(longi_w,longi_e), time=time_NH).compute()
pr_avg
```

```python
day = np.arange(-15,16,1)
nlag = len(day)
lag_xr = xr.DataArray(day, coords=[day], dims=['day'])
lag_xr
```

```python
lon = pr_avg['lon']
lat = pr_avg['lat']
nlon = len(lon)
nlat = len(lat)
nlon
```

```python
kw_lag_avg = np.zeros((nlag,nlon))
kw_lag_std = np.zeros((nlag,nlon))

pr_kelvin_lreg_NH = np.zeros((nlat,nlon,nlag))
pr_kelvin_corr_NH = np.zeros((nlat,nlon,nlag))
pr_kelvin_lreg_NH.shape
```

```python
for t in range(0,nlag,1):
    lag = time_NH+np.timedelta64(int(lag_xr[t].values),'D')
    kw_lag = data_kw['precipitationCal'].sel(time=lag, lat=slice(latitude_s,latitude_n),lon=slice(longi_w,longi_e))
    kw_lag_avg[t,:] = kw_lag.mean(dim='lat').mean(dim='time')
    kw_lag_std[t,:] = kw_lag.mean(dim='lat').std(dim='time')
```

```python
#create xarray

pr_kelvin_lag_avg_xr = xr.DataArray(kw_lag_avg, coords=[day, lon], dims=('day','lon'))
pr_kelvin_lag_std_xr = xr.DataArray(kw_lag_std, coords=[day, lon], dims=('day','lon'))
```

```python
dtime = 0.125
lags_subdaily = np.arange(-15,15+dtime,dtime)
```

```python
kw_lag_avg_interp = pr_kelvin_lag_avg_xr.interp(day=lags_subdaily,method='cubic')
kw_lag_std_interp = pr_kelvin_lag_std_xr.interp(day=lags_subdaily,method='cubic')

day_subdaily = kw_lag_avg_interp['day']
```

```python
pr_kelvin_lag_avg_xr.name = "precipitationCal"
pr_kelvin_lag_std_xr.name = "precipitationCal"
```

```python
# t-val

t_val = (kw_lag_avg_interp/kw_lag_std_interp) * (((len(time_NH)-1))**(1/2))
```

```python
#p-val

p_val = scipy.stats.t.sf(abs(t_val), df=len(time_NH)-1)*2
p_val_xr = xr.DataArray(p_val, coords=[day_subdaily, lon], dims=('day','lon'))
```

```python
p_val_true = kw_lag_avg_interp.where(p_val<= 0.01)
```

```python
p_val_true.name = "precipitationCal"
```

```python
dro = '/home/ffahrin/test/IMERG/rolling_avg/'
fn_output = wave+'_NH_base_strong_pr.nc'
    
# save netcdf 
p_val_true.to_netcdf(path=dro+fn_output)
```

```python
#load ITCZ state data

dri = '/home/ffahrin/test/IMERG/ITCZ_states/'
fn  = 'IMERG_ITCZ_st_daily_EastPac_60prc.nc'

data_itcz = xr.open_dataset(dri+fn)
```

```python
itcz_new_xr = data_itcz['itcz_state'].sel(time=time_fma)
itcz_new_xr
```

```python
cond_dITCZ = (itcz_new_xr == 1)
cond_nITCZ = (itcz_new_xr == 2)
cond_sITCZ = (itcz_new_xr == 3)
cond_aITCZ = (itcz_new_xr == 4)
cond_eITCZ = (itcz_new_xr == 5)
```

```python
prc_dITCZ = round(len(itcz_new_xr.where((cond_dITCZ), drop=True))/(len(time_fma))*100,1)
prc_nITCZ = round(len(itcz_new_xr.where((cond_nITCZ), drop=True))/(len(time_fma))*100,1)
prc_sITCZ = round(len(itcz_new_xr.where((cond_sITCZ), drop=True))/(len(time_fma))*100,1)
prc_aITCZ = round(len(itcz_new_xr.where((cond_aITCZ), drop=True))/(len(time_fma))*100,1)
prc_eITCZ = round(len(itcz_new_xr.where((cond_eITCZ), drop=True))/(len(time_fma))*100,1)
```

```python
print('Double     ITCZ happens '+str(prc_dITCZ)+'% of the time.')
print('Northern   ITCZ happens '+str(prc_nITCZ)+'% of the time.')
print('Southern   ITCZ happens '+str(prc_sITCZ)+'% of the time.')
print('Absent     ITCZ happens '+str(prc_aITCZ)+'% of the time.')
print('Equatorial ITCZ happens '+str(prc_eITCZ)+'% of the time.')
```
